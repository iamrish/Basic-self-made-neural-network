```python
import pandas as pd
import numpy as np
import scipy as sc
from sklearn import model_selection as ms
```

```python
y = data["Survived"]
data = data.drop("Survived", axis=1)
```

```python
data_np = data.to_numpy()
# data_np.shape
y = y.to_numpy()
data_np[1]
```

```python
data_train, data_test = ms.train_test_split(data_np, train_size=0.8)
```

```python
y, y_test = ms.train_test_split(y, train_size=0.8)
```

```python
data_train.shape[0]
```

```python
class neuron1:
    
    def __init__(self):
        self.weights = np.random.randn(7,1)
        self.bias = np.random.randn()
```

```python
class neuron2:
    
    def __init__(self):
        self.weights = np.random.randn(3,1)
        self.bias = np.random.randn()
```

```python
neuron_11 = neuron1()
neuron_12 = neuron1()
neuron_13 = neuron1()
neuron_21 = neuron2()
```

```python
print("w_11: {}".format(neuron_11.weights))
print("b_11: {}".format(neuron_11.bias))
print("w_12: {}".format(neuron_12.weights))
print("b_12: {}".format(neuron_12.bias))
print("w_13: {}".format(neuron_13.weights))
print("b_13: {}".format(neuron_13.bias))
print("w_21: {}".format(neuron_21.weights.shape))
print("b_21: {}".format(neuron_21.bias))
```

```python
def forward_prop(neu, X):
    
#     print(neu.weights.shape)
#     print(X.shape)

    z = np.matmul(neu.weights.T, X.T) + neu.bias
    z = np.array(z, dtype=np.float128)
    a = 1/(1 + np.exp(-z))
    return a
```

```python
def cost_function(y_hat, i):
    
    cost = -(y[i]*np.log(y_hat)) - ((1-y[i])*np.log(1-y_hat))
    return cost
```

```python

alpha = 0.8

for a in range(1000):
    loss = 0
    for i in range(data_train.shape[0]):
        a_11 = forward_prop(neuron_11, data_train[i].reshape(1, -1))
        a_12 = forward_prop(neuron_12, data_train[i].reshape(1, -1))
        a_13 = forward_prop(neuron_13, data_train[i].reshape(1, -1))
        temp = np.array([a_11[0][0], a_12[0][0], a_13[0][0]])
        temp = temp.reshape(1,3)
        a_21 = forward_prop(neuron_21, temp.reshape(1, -1))
        y_hat = a_21[0][0]
        cost = cost_function(a_21, i)
        loss = loss+cost
        a_1 = np.array([a_11[0][0], a_12[0][0], a_13[0][0]])
        a_1 = a_1.reshape(3, 1)

        # dcost/dy_hat
        dy_hat = (y_hat-y[i])/((y_hat)*(1-y_hat))

        # dcost/dz_21
        dz_21 = y_hat-y[i]

        # dcost/dneuron_21.weights
        dneuron_21_weights = a_1*dz_21

        # dcost/dneuron_21.bias
        dneuron_21_bias = dz_21

        # dcost/da_1i for all i = {1,2,3}
        da_1 = neuron_21.weights*dz_21

        # dcost/dz_1i for all i = {1, 2, 3}\
        dz_1 = []
        for j in range(len(a_1)):
            dz_1.append(dz_21*neuron_21.weights[j][0]*a_1[j]*(1-a_1[j]))
        dz_1 = np.array(dz_1)

        # dcost/dneuron_1i.weights for all i = {1, 2, 3}
        dneuron_11_weights = data_train[i]*dz_1[0][0]
        dneuron_12_weights = data_train[i]*dz_1[1][0]
        dneuron_13_weights = data_train[i]*dz_1[2][0]

        dneuron_11_weights = dneuron_11_weights.reshape(1, -1)
        dneuron_12_weights = dneuron_12_weights.reshape(1, -1)
        dneuron_13_weights = dneuron_13_weights.reshape(1, -1)

        # dcost/dbias_1i.bias for all i={1, 2, 3}
        dneuron_11_bias = dz_1[0][0]
        dneuron_12_bias = dz_1[1][0]
        dneuron_13_bias = dz_1[2][0]

        # updating weights and biases
        neuron_11.weights = neuron_11.weights - alpha*dneuron_11_weights.T
        neuron_12.weights = neuron_12.weights - alpha*dneuron_12_weights.T
        neuron_13.weights = neuron_13.weights - alpha*dneuron_13_weights.T

        neuron_11.bias = neuron_11.bias - alpha*dneuron_11_bias
        neuron_12.bias = neuron_12.bias - alpha*dneuron_12_bias
        neuron_13.bias = neuron_13.bias - alpha*dneuron_13_bias

        neuron_21.weights = neuron_21.weights - alpha*dneuron_21_weights
        neuron_21.bias = neuron_21.bias - alpha*dneuron_21_bias 
    loss = loss/data_train.shape[0]
    print("epoch {}: {}".format(a, loss))
```

```python
print("w_11: {}".format(neuron_11.weights))
print("b_11: {}".format(neuron_11.bias))
print("w_12: {}".format(neuron_12.weights))
print("b_12: {}".format(neuron_12.bias))
print("w_13: {}".format(neuron_13.weights))
print("b_13: {}".format(neuron_13.bias))
print("w_21: {}".format(neuron_21.weights))
print("b_21: {}".format(neuron_21.bias))
```

```python
# Testing
count = 0
for i in range(data_test.shape[0]):
    a_11 = forward_prop(neuron_11, data_test[i].reshape(1, -1))
    a_12 = forward_prop(neuron_12, data_test[i].reshape(1, -1))
    a_13 = forward_prop(neuron_13, data_test[i].reshape(1, -1))
    temp = np.array([a_11[0][0], a_12[0][0], a_13[0][0]])
    temp = temp.reshape(1,3)
    a_21 = forward_prop(neuron_21, temp.reshape(1, -1))
    y_hat = a_21[0][0]
    
    if y_hat>=0.5:
        y_hat = 1
    else:
        y_hat = 0
    
    if(y_hat == y_test[i]):
        count = count + 1
        
print(count/data_test.shape[0])    
```
